```python
'''26. Matrix transpose using list comprehension
    Input: [[1, 2, 3], [4, 5, 6]]
    Output: [[1, 4], [2, 5], [3, 6]]'''
matrix = [[1, 2, 3], [4, 5, 6]]

transpose = [[row[i] for row in matrix] for i in range(len(matrix[0]))]

print(transpose)
```

```python
'''27. Flatten 3D list
    Input: [[[1, 2], [3]], [[4, 5]]]
    Output: [1, 2, 3, 4, 5]'''
Input: [[[1, 2], [3]], [[4, 5]]]
data = [[[1, 2], [3]], [[4, 5]]]

flat = [num for sublist1 in data for sublist2 in sublist1 for num in sublist2]

print(flat)
```

```python
'''30. Generate list of all substrings of a string
    Input: "abc"
    Output: ["a", "ab", "abc", "b", "bc", "c"]'''
s = "abc"

substrings = [s[i:j] for i in range(len(s)) for j in range(i+1, len(s)+1)]

print(substrings)

```
